```python
data.columns = data.columns.str.lower().str.replace(' ','_')

```

```python
data.head()
```

Age Column

```python
med_age = data.age.mean()
print(med_age)
```

```python
data['age'] = data['age'].fillna(med_age)
data['age'] = data['age'].round(decimals=1)
data['age']
```

```python
data.isna().sum()
```

Salary Column

```python
data['salary'] = data['salary'].apply(lambda x:x.replace('$',''))
data['salary'] = data['salary'].apply(lambda x:x.replace('k','000'))
```

```python
data['salary']
```

```python
data.head()
```

```python
data['age'] = data['age'].astype(int)
```

Rating Column

```python
rat_av = data.rating.mean()
print(rat_av)
```

```python
data['rating'] = data['rating'].fillna(rat_av)
data['rating'] = data['rating'].round(decimals=1)
data['rating']
```

```python
data.isna().sum()
```

```python
data.replace(-1,1,inplace = True)
```

```python
data.head()
```

```python
data['rating']
```

Location Column

```python
data['location']
```

```python
data['location'] = data['location'].str.replace(',In','')
data['location'] = data['location'].str.replace(',Ny','')
data['location'] = data['location'].str.replace(' In','')
data['location'] = data['location'].str.replace(' Aus','')
data['location']
```

```python
data['location'] = data['location'].astype(str)
```

Established Column

```python
data["established"] = data["established"].replace(-1,np.nan)
data["established"].fillna('Unknown', inplace=True)
```

```python
data.head()
```

Easy_Apply Column


```python
data['easy_apply'] = data['easy_apply'].replace("-1","TRUE")
```
